```python
df['team_a_form'] = df['team_a_form'].fillna(df['team_a_form'].median())
df['injured_a'] = df['injured_a'].fillna(df['injured_a'].median())
df['attack_a'] = df['attack_a'].fillna(df['attack_a'].median())
df.isnull().sum()
```

```python
df['match_date'] = pd.to_datetime(df['match_date'], errors='coerce')
df.info()
```

```python
df['rank_diff'] = df['team_b_rank'] - df['team_a_rank']

df['attack_diff'] = df['attack_a'] - df['attack_b']

df['defense_diff'] = df['defense_a'] - df['defense_b']

df['goal_diff_last5'] = (
    df['goals_scored_last5_a']
    - df['goals_scored_last5_b']
)

df['injury_diff'] = df['injured_b'] - df['injured_a']
df['abs_rank_diff'] = abs(df['team_a_rank'] - df['team_b_rank'])
df['abs_attack_diff'] = abs(df['attack_a'] - df['attack_b'])
df['abs_defense_diff'] = abs(df['defense_a'] - df['defense_b'])
df['abs_form_diff'] = abs(df['team_a_form'] - df['team_b_form'])

df['abs_goal_diff'] = abs(
    df['goals_scored_last5_a'] -
    df['goals_scored_last5_b']
)
df['similar_strength'] = (
    (df['abs_rank_diff'] <= 5) &
    (df['abs_attack_diff'] <= 5) &
    (df['abs_defense_diff'] <= 5)
).astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
```

```python
X = df.drop(["match_outcome", "match_date"], axis=1)
y = df["match_outcome"]
```

```python
num_cols = X.select_dtypes(include=["int64","float64"]).columns
```

```python
preprocessor = ColumnTransformer([
        ("num", StandardScaler(), num_cols)
    ]
)
preprocessor
```

```python
lr_model = Pipeline([
    ("prep", preprocessor),
    ("clf", LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42))
])
rf_model = Pipeline([
    ('smote', SMOTE(random_state=42)),
    ("prep", preprocessor),
    ("clf", RandomForestClassifier(n_estimators=300,class_weight='balanced',random_state=42))
])
```
